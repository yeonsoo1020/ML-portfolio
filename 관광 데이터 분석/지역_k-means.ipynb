```python
def area(x):
  if x == '서울특별시':
    return '서울'
  elif x == '경기도' :
    return '경기'
  elif x == '인천광역시':
    return '인천'
  elif x == '강원도':
    return '강원'
  elif x == '제주특별자치도':
    return '제주'
  elif x == '경상북도':
    return '경북'
  elif x == '전라남도':
    return '전남'
  elif x == '부산광역시':
    return '부산'
  elif x == '전라북도':
    return '전북'
  elif x == '충청남도':
    return '충남'
  elif x == '경상남도':
    return '경남'
  elif x == '충청북도':
    return '충북'
  elif x == '대구광역시':
    return '대구'
  elif x == '대전광역시':
    return '대전'
  elif x == '광주광역시':
    return'광주'
  elif x == '울산광역시':
    return '울산'
  elif x == '세종특별자치시':
    return '세종'
  else:
    return x
```

```python
df['가맹시'] = df['가맹시'].apply(area)
```

```python
df['가맹시'].value_counts()
```

```python
df.iloc[14160, 9] = '서울특별시 강남구 언주로 126길 25'
```

```python
df.iloc[3035, 9] = '인천광역시 강화군 남문안길 7'
```

```python
df['가맹구'] = df['ADDR'].str.split(' ').str[1]
```

```python
def area2(x):
  if (x == '조치원읍') | (x == '전동면') | (x == '금남면') | (x == '부강면') | (x == '도담동') | (x == '호수공원길') |  (x == '한누리대로') | (x == '다솜로') :
    return '세종'
  else:
    return x
```

```python
df['가맹구'] = df['가맹구'].apply(area2)
```

```python
df['지역'] = df['가맹시'] + " " + df['가맹구']
```

```python
df['지역'] = df['지역'].apply(lambda x : '인천 미추홀구' if x == '인천 남구' else x)
```

```python
df = df[(df['PLACE_TY'] == 'restaurant') | (df['PLACE_TY'] == 'playground') | (df['PLACE_TY'] == 'cafe')].reset_index(drop = True)
```

```python
df['장소분류'] = df['PLACE_TY'] + ' ' + df['MEDIA_TY']
```

```python
df = pd.get_dummies(df, columns = ['MEDIA_TY', '장소분류'])
```

# k-means

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
# 군집에 사용할 변수 선택
data4 = data6[['지역', '결제액_쇼핑', '결제액_음식', '결제액_체험','영화장소', '쇼장소', '드라마장소',
       '아티스트장소', '관광지수', '음식점수', '쇼핑센터수', '음식종류_외국음식', '음식종류_한식']]
```

```python
data4.head(2)
```

```python
#표준화
sc = StandardScaler()
df_scaled = sc.fit_transform(data4.iloc[:, 1:])
pd.DataFrame(df_scaled)
```

```python
import matplotlib.pyplot as plt

ks = range(1,15)
inertias = []

for k in ks:
  model = KMeans(n_clusters=k)
  model.fit(df_scaled)
  inertias.append(model.inertia_)

# Plot ks vs inertias
plt.plot(ks, inertias, '-o')
plt.xlabel('number of clusters, k')
plt.ylabel('inertia')
plt.xticks(ks)
plt.show()
```

## (1) k = 3

```python
# 클러스터링
kmeans = KMeans(n_clusters=3, random_state=0)
clusters = kmeans.fit(df_scaled)

# 클러스터링 변수
data4['cluster'] = clusters.labels_
data4.head(2)
```

```python
# 그룹별 중앙값
data4.groupby('cluster').median()
```

```python
data4.groupby('cluster').mean()
```
